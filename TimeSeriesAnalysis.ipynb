## TIME SERIES FORECASTING

## Data Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
df = pd.DataFrame()
df['Date'] = pd.date_range(start='01/01/2019', end='12/30/2021')
df['Sales'] = np.random.randint(50000, 100000, size=(len(df['Date']), 1))
print(df['Date'])
print(df['Sales'])
# df.to_csv('Sales.csv', index=False)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y') 
df = df.set_index('Date')
```

```python
df.head()
```

## Visualize Time Series

```python
df=df['Sales']
plt.figure(figsize=(15,10))
df.plot() 
plt.legend()
```

## Determining Stationarity

## With Rolling Statistics

```python
rolling_mean = df.rolling(window=365).mean() 
rolling_std = df.rolling(window=365).std() 
plt.figure(figsize=(20,10)) 
plt.plot(df, color = 'green', label = 'Original') 
plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean') 
plt.plot(rolling_std, color = 'magenta', label = 'Rolling Std') 
plt.legend(loc = 'best') 
plt.title('Rolling Mean & Rolling Standard Deviation') 
plt.xlabel('Date') 
plt.ylabel('Sales') 
plt.legend()
```

```python
#Conclusion From Rolling Statistics: Stationary Data Because Of Constant Mean And Sd
```

## With Augmented Dickey Fuller Test

```python
from statsmodels.tsa.stattools import adfuller 
adf_test = adfuller(df) 
print('ADF Statistics: {}'.format(adf_test[0])) 
print('p-value: {}'.format(adf_test[1])) 
for key, value in adf_test[4].items(): 
    print('Critical Values Over {}: {}'.format(key, value)) 
```

```python
#Conclusion From ADF Test: Stationary Data Because p's value is less than 0.005 and ADF statistics is less than critical values
```

## Time Series Decomposition

```python
import statsmodels.api as sm
ts_decomp = sm.tsa.seasonal_decompose(X_train, model='additive')
ts_decomp.plot()
plt.show()
```

```python
# Since, we have seasonality in the data, we are going to use SARIMA model in place of ARIMA. SARIMA have 7 components, 3 of ARIMA model and 4 of seasonal component
```

```python
import itertools
import warnings
warnings.filterwarnings("ignore")
p = d = q = range(0,2)

pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
metric_aic_dict=dict()

for pm in pdq:
    for pm_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(X_train, order=pm, seasonal_order=pm_seasonal, enforce_stationarity=False, enforce_invertibility=False)
            model_aic = model.fit()
            print('ARIMA{}*{}12 - AIC:{}'.format(pm, pm_seasonal, model_aic.aic))
            metric_aic_dict.update({(pm, pm_seasonal):model_aic.aic})
        except:
            continue
            
```

```python
{k: v for k, v in sorted(metric_aic_dict.items(), key=lambda x: x[1])}
```

## Fitting SARIMA Model

```python
model = sm.tsa.statespace.SARIMAX(X_train, order=(1, 0, 1), seasonal_order=(1, 1, 1, 12), enforce_stationarity=False, enforce_invertibility=False)
model_aic = model.fit()
model_aic.summary()
```

```python
forecast = model_aic.get_prediction(pd.to_datetime('2021-12-01'), dynamic=False)
predictions = forecast.predicted_mean
predictions
```

## Calculating the RMSE

```python
actual = X_test['2021-12-01':]
rmse = np.sqrt((predictions - actual) ** 2).mean()
rmse
```
